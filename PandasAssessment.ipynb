### Q2. You are given a list of student names and their scores. Convert it into a Series and display the score of the student named "Anu".

```python
names = ["Ravi", "Anu", "Simran", "Raj", "Priya"]
scores = [85, 92, 78, 88, 90]
# Create a Series with names as index and scores as values
student_scores = pd.Series(scores, index=names)

# Display Anu's score
print(student_scores["Anu"])
#student_scores["Anu"] # or student_scores[1] # output - 92
```

Why Both Work:
1. student_scores[1]
This uses positional indexing.

Since "Anu" is at the 2nd position (index 1), this returns 92.

---

2. student_scores["Anu"]
This uses label-based indexing.

It directly accesses the score by using "Anu" as the index label.

### Q4. From a DataFrame of employees with columns Name, Salary, and Department, use .loc to display the salary of the employee named "Raj".

```python
d=[
    {"Name":"Divija","Salary":20000,"Department":"Software Engineer"},
    {"Name":"Tanvi","Salary":30000,"Department":"VLSI Designer"},
    {"Name":"Vrinda","Salary":10000,"Department":"Data Analyst"},
    {"Name":"Raj","Salary":10000,"Department":"UI/UX Designer"}
]

df= pd.DataFrame(d)
df
```

```python
df.loc[3,"Salary"]
```

```python
# Use .loc to display the salary of "Raj"
salary_raj = df.loc[df["Name"] == "Raj", "Salary"] # Step 1: df["Name"] == "Raj" , It checks every row: “Is the name Raj?” → Only the last row is True.
# Step 2: df.loc[df["Name"] == "Raj", "Salary"] here .loc[<rows>, <columns>] ka format hota hai.
# This means: Select row(s) where Name == "Raj" (i.e., row 3).
# From that row, select the "Salary" column only.
print(salary_raj.values[0]) # .values converts the Series into a numpy array → [10000], and [0] gives the first value.
# That means: .values → [10000] and [0] → first value = 10000
```

Think of this:

list = [10, 20, 30]

print(list[1])  # prints 20

---

Similarly:

salary_raj.values = [10000]

salary_raj.values[0] = 10000

### Q7. Create a DataFrame with some NaN values in the Marks column. Replace all NaN values with the average (mean) of the Marks column.

```python
data = {
    "Name": ["Divija", "Tanvi", "Vrinda", "Raj", "Simran"],
    "Marks": [99, np.nan, 90, np.nan, 95] #NaN values in the marks column
}

df = pd.DataFrame(data)
df
```

```python
df["Marks"]
```

```python
mean_marks = df["Marks"].mean() #  Calculate the mean of the Marks column (ignoring NaN)
# Replace NaN with the mean
df["Marks"].fillna(mean_marks, inplace=True) # Fill all NaN values with mean_marks.
df["Marks"]

#fillna() is a Pandas function used to replace missing values (NaN) with something you provide (like a number, string, method, etc).

#  inplace=True
# This tells Pandas: "Do the change directly in the original DataFrame."
# If we don’t write inplace=True, it will just return a modified copy, and our original DataFrame won't change unless we assign it back.
```

### Q11. You are given a DataFrame df with missing values in the Score column. Use .loc to fill only those missing values with 75.

```python
data = {
    "Name": ["Divija", "Tanvi", "Vrinda", "Raj"],
    "Score": [90, np.nan, 85, np.nan]
}

df = pd.DataFrame(data)
df
```

```python
df['Score']
```

```python
marks = 75 #  assigning 75 to marks
# Replace NaN with the marks
df["Score"].fillna(marks, inplace=True) # Fill all NaN values with mean_marks.
df["Score"]
```

```python
# Use .loc to fill only those missing values with 75.
data = {
    "Name": ["Divija", "Tanvi", "Vrinda", "Raj"],
    "Score": [90, np.nan, 85, np.nan]
}
df = pd.DataFrame(data)

df.loc[df["Score"].isnull(), "Score"] = 75 # df["Score"].isnull() → returns True for rows where Score is NaN.
# df.loc[condition, "Score"] = 75 → updates only those cells where the condition is True.
df
```

### Q15. You have a DataFrame of city temperatures. Some values in the Temperature column are missing. Fill them using the average temperature calculated using mean().

```python
data = {
    "City": ["Delhi", "Mumbai", "Chennai", "Kolkata", "Bangalore"],
    "Temperature": [35.5, np.nan, 33.0, np.nan, 29.5]
}

df = pd.DataFrame(data)
df
```

```python
mean_temp = df["Temperature"].mean() # Calculating the average temperature (excluding NaN)

df["Temperature"].fillna(mean_temp, inplace=True) # Fill missing values with the mean

df
```
