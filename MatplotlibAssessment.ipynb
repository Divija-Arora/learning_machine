### 8. Create a pie chart with percentage labels: Subjects = ['Math', 'Science', 'English'], Values = [30, 40, 30]

```python
# The percentage labels in a Matplotlib pie chart are added using the autopct parameter in the plt.pie() function.
subjects = ['Math', 'Science', 'English']
values = [30, 40, 30]
# Create pie chart
plt.pie(values, labels=subjects, autopct='%1.1f%%') # autopct stands for “automatic percentage”
#'%1.1f%%' tells Matplotlib to:
# Show the percentage with 1 decimal place (e.g., 33.3%)
# The double % (%%) is needed to print a literal % symbol in the output.
plt.show()
```

### 11. Create a subplot with two plots: one line plot and one bar chart.

```python
x = [1, 2, 3, 4]
y_line = [10, 20, 25, 30]
y_bar = [5, 15, 20, 25]

# Create a subplot with 1 row and 2 columns
fig, axs = plt.subplots(1, 2, figsize=(10, 4))  # 1 row, 2 columns
# axs stands for axes — a list (or array) of individual plots/subplots.
# In this case, axs[0] is the left plot, and axs[1] is the right plot.
# figsize=(10, 4) make the total figure 10 inches wide and 4 inches tall.

# Line plot on the first subplot
axs[0].plot(x, y_line, color='blue', marker='o')
axs[0].set_title('Line Plot')
axs[0].set_xlabel('X-axis')
axs[0].set_ylabel('Y-axis')

# Bar chart on the second subplot
axs[1].bar(x, y_bar, color='green')
axs[1].set_title('Bar Chart')
axs[1].set_xlabel('X-axis')
axs[1].set_ylabel('Y-axis')

plt.show()
```

### 16. Plot a line graph and annotate the highest point on it.

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([10, 25, 18, 30, 22])

plt.plot(x, y, marker='o')

max_y = max(y)
max_x = x[y.argmax()]  # Get corresponding x-value

# Annotate the highest point
plt.annotate( # plt.annotate() places a label with an arrow pointing to that point.
    'Highest Point',
    xy=(max_x, max_y),             # Point to annotate
    xytext=(max_x + 0.2, max_y + 2), # Position of the label
    arrowprops=dict(facecolor='red', shrink=0.05)
)
plt.show()
```

1. 'Highest Point' - Ye text us point ke upar likha jayega.
---

2. xy=(max_x, max_y) - Ye vo point hai graph ke upar, jahan arrow point karega.
Means: Kis point ko highlight karna hai? (e.g. max point)
---

3. xytext=(max_x + 0.2, max_y + 2)- Ye text kahaan dikhana hai, arrow ke side mein.

I moved it a bit so it doesn’t overlap the point.
(Thoda right aur upar kar diya text)

---
4. arrowprops=dict(facecolor='red', shrink=0.05)

facecolor='red' → arrow ka color red hoga

shrink=0.05 → arrow thoda chhota ho jata hai

### 17. Create a bar chart and add value labels on top of each bar.

```python
x=np.array(["Sachin","Virat","Viv-Richard","Sunil Gavaskar"])
y=np.array([18000,15000,7500,9000])
plt.bar(x,y)
for i in range(len(x)):
    plt.text(i, y[i] + 200, str(y[i]), ha='center', fontsize=10)
plt.show()
# plt.text(i, y[i] + 200, str(y[i]), ha='center')
# i → x-position of the bar, y[i] + 200 → little above the top of the bar
# str(y[i]) → the label to show (run value), ha='center' → horizontally center the text above the bar
```

### 18. Create a plot with x-axis rotated 45 degrees using tick labels: Months = ['Jan', 'Feb', 'Mar', 'Apr'], Sales = [250, 300, 150, 400]

```python
months = ['Jan', 'Feb', 'Mar', 'Apr']
sales = [250, 300, 150, 400]

plt.bar(months, sales)

# Rotate x-axis labels
plt.xticks(rotation=45)

# Labels and title
plt.xlabel("Months")
plt.ylabel("Sales")

plt.show()
# why ticks are not sales here? and why months are ticks - jan, feb...
# Because sales are data values being plotted against the ticks (months).
# Tick labels are text on the axis to help you read the graph. Sales are the data values, not labels — so they’re not "ticks".
```

### 19. Draw three subplots in a single row, each with a different plot (line, bar, pie).
